# Дана функция f(x) = 5x^2 + 10x - 30
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
a = 5
b = 10
c = -30

x = Symbol("x")
f_x = a*(x**2)+b*(x**1)+c
```

```python
print(f'Корень №1: {solve(f_x,x)[0].n()}\nКорень №2: {solve(f_x,x)[1].n()}')
```

```python
from sympy.plotting import plot

g1 = plot(f_x,(x,-5,3), line_color='red', title='график функции')
```

```python
derivative = diff(f_x)
derivative
```

```python
g2 = plot(derivative,(x,-5,3), line_color='red', title='график функции')
```

```python
g3 = plot(f_x, derivative,(x,-5,3), title='Совмещенный график', legend=True, xlabel='X', ylabel='Y', show=False)
g3[0].line_color = 'red'
g3[1].line_color = 'green'
g3.show()
```

```python
print('Корень №1: ', solve(derivative,x)[0].n(3))
```

```python
solve(derivative > 0, x)
```

```python
solve(derivative < 0, x)
```
